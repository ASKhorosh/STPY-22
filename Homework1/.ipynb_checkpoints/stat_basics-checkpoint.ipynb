```python
hd['pulse'] = hd['pulse'].astype(float)
```

```python
#выбросы по пульсу
q1 = hd['pulse'].quantile(0.25)
q3 = hd['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = hd[hd['pulse'].between(lower_bound, upper_bound, inclusive=True)]
remove_outliers
```

```python
hd['rectal temperature'] = hd['rectal temperature'].astype(float)
```

```python
#выбросы по температуре
q1 = hd['rectal temperature'].quantile(0.25)
q3 = hd['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = hd[hd['rectal temperature'].between(lower_bound, upper_bound, inclusive=True)]
remove_outliers
```

```python
#Пропуски в столбце боли можно заполнить по данным столбца исхода, предполагая, что есть зависимость его от степени боли
print(hd.groupby('outcome')['pain'].median())
```

```python
fill_median_by_outcome = hd
fill_median_by_outcome['pain'].fillna(hd.groupby('outcome')['pain'].transform('median'), inplace=True)
```

```python
#Пропуски в столбцах температур и пульса можно заполнить, группируя по боли. Т.к. степень боли может определяться как раз этим
print(hd.groupby('pain')['rectal temperature'].median())
```
