### 1) Как критики относятся к спортивным играм? 
Проверим, если критики ставят оценку играм больше 70, значит они хорошо относятся к ним, иначе недостаточно хорошо. 

H0: средняя удовлетворенность критиков спортивными играми <= 70
H1: средняя удовлетворенность критиков спортивными играми 70

alpha = 0.05

```python
#офильтруем спортивные игры
sports = df.loc[(df['Genre'] == 'Sports')&(df['Critic_Score'] > 0)]
sports
```

```python
from scipy import stats as st

alpha = 0.05
result = st.ttest_1samp(sports['Critic_Score'], 70, alternative='greater')

print(sports['Critic_Score'].mean())
print(result)

if result.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу, средняя оценка больше 70')
else:
    print('Не отвергаем нулевую нулевую гипотезу, средняя оценка меньше 70')
```

p-value меньше 5% - мы получили значимый результат. p-value меньше уровня значимости, нулевую гипотезу отвергаем. это значит, что критики относятся к спортивным играм достаточно хорошо: оценка 70 и больше из 100

### 2) Критикам нравятся больше игры на PC или на PS4?

Выясняем, есть ли разница между средними значениями оценок игр на РС или PS4? Если разница значима, сравним средние значения и выясним, что нравится больше.

H0: средние оценки одинаковые
H1: средние оценки различаются

alpha = 0.05

```python
#готовим выборки
PC = df.loc[(df['Platform'] == 'PC')&(df['Critic_Score'] > 0)]
PC
```

```python
PS4 = df.loc[(df['Platform'] == 'PS4')&(df['Critic_Score'] > 0)]
PS4
```

```python
alpha = 0.05
result = st.ttest_ind(PC['Critic_Score'], PS4['Critic_Score'], equal_var=False)

print('PC', PC['Critic_Score'].mean())
print('PS4', PS4['Critic_Score'].mean())

print(result)

if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по разному относятся к платформам')
else:
    print('Не отвергаем нулевую гипотезу, у критиков есть предпочтения между двумя платформами')
```

p-value меньше 5% - мы получили значимый результат. p-value меньше уровня значимости, нулевую гипотезу отвергаем. это значит, что критики относятся к платформам не одинаково. сравним средние оценки и узнаем, что нравится больше

```python
if (PC['Critic_Score'].mean() < PS4['Critic_Score'].mean()):
    print('Критикам больше нравятся игры на PS4')
else:
    print('Критикам больше нравятся игры на PC')
```

### 3) Критикам больше нравятся стрелялки или стратегии?

Выясняем, есть ли разница между средними значениями оценок стрелялок или стратегий? Если разница значима, сравним средние значения и выясним, что нравится больше.

H0: средние оценки одинаковые H1: средние оценки различаются

alpha = 0.05

```python
#готовим выборки
shooter = df.loc[(df['Genre'] == 'Shooter')&(df['Critic_Score'] > 0)]
shooter
```

```python
strategy = df.loc[(df['Genre'] == 'Strategy')&(df['Critic_Score'] > 0)]
strategy
```

```python
alpha = 0.05
result = st.ttest_ind(shooter['Critic_Score'], strategy['Critic_Score'], equal_var=False)

print('Shooter', shooter['Critic_Score'].mean())
print('Strategy', strategy['Critic_Score'].mean())

print(result)

if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по разному относятся к выбранным жанрам')
else:
    print('Не отвергаем нулевую гипотезу, у критиков есть предпочтения между двумя жанрами')
```

p-value меньше 5% - мы получили значимый результат. p-value меньше уровня значимости, нулевую гипотезу отвергаем. это значит, что критики относятся к жанрам не одинаково. сравним средние оценки и узнаем, что нравится больше

```python
if (shooter['Critic_Score'].mean() < strategy['Critic_Score'].mean()):
    print('Критикам больше нравятся стратегии')
else:
    print('Критикам больше нравятся стрелялки')
```

```python
import glob
import re
```

```python
spam = pd.read_csv('spam.csv')
spam
```

```python
# 1. приведем все слова к нижнему регистру

 = spam['Message'].str.lower()
spam
```

```python
# 2. уберем все служебные символы

spam['Message'] = [re.sub('[\W_]+',' ', text) for text in spam['Message']]
spam
```

```python
import nltk
nltk.download('stopwords')
```

```python
#3. удалим стоп-слова
from nltk.corpus import stopwords
stop = stopwords.words('english')

spam['Message'].apply(lambda x: [item for item in x if item not in stop])
spam
```

```python
import nltk
nltk.download('wordnet')
```

```python
#4. делаем лемматизацию

from nltk.stem import WordNetLemmatizer

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text): #функция лемматизации входного текста
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]

spam['message_lemmatized'] = spam.Message.apply(lemmatize_text) #применяем функцию для датафрейма
spam
```

```python
import string

def list_to_str(text): #функция возвращения строки из списка
        return ' '.join(text)

spam['message_lemmatized'] = [list_to_str(l) for l in spam['message_lemmatized']]
spam
```

```python
#5. преобразование в вектора

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(spam['message_lemmatized'])
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
tfidf_matrix
```

```python
#6-7. делим данные на тестовые и тренировочные и строим модель

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split

X = spam['message_lemmatized']
y = spam['Category']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
#оцениваем точность

y_pred = logreg.predict(X_test)
print('Точность на тестовой выборке:'.format(logreg.score(X_test, y_test)))
```

```python
#8. матрица confusion_matrix

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
confusion_matrix
```

```python
#9. строим датафрейм

spam['Category_pred'] = spam.Category.apply(y_pred)
spam_filtered = spam.loc[(spam['Category'] != spam['Category_pred'])]
spam_filtered
```
